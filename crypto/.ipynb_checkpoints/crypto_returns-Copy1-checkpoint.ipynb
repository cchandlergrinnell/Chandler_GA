```python
def data_creater(tics):
    end = datetime.now()
    start = end - timedelta(hours=2000)
    df_curr = pd.DataFrame(index=pd.date_range(start, end, freq='H'))
    for x in tics:
        fysm = x
        tsym = 'BTC'
        d = json.loads(requests.get('https://min-api.cryptocompare.com/data/histohour?fsym=%s&tsym=%s&limit=6000&aggregate=1' % (fysm ,tsym)).text)
        json_normalize(d)
        df = pd.DataFrame(d['Data'])
        df['time'] = pd.to_datetime(df['time'])
        end = datetime.now()
        start = end - timedelta(hours=(len(df.index)-1))
        df['time'] = pd.date_range(start, end, freq='H')
        pd.merge(df_curr, df['close'], right_on='time')
    return df_curr
```
