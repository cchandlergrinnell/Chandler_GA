```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_excel("crypto_ret_1.xlsx", index_col='time')
```

```python
#convert to returns by shifting
ret = df.pct_change(24, fill_method=None)
ret = ret.replace([np.inf, -np.inf], np.nan)
ret.columns.tolist()
```

```python
analysis = ['NEO', 'GAS', 'ETH', 'MCO', 'DASH', 'LTC', 'BTC']
ret[analysis].plot(legend=True)
```
