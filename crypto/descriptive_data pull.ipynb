```python
import pandas as pd
import json
from bs4 import BeautifulSoup
import requests
```

### Coin DataFrame

```python
url = "https://www.cryptocompare.com/api/data/coinlist/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
dic = json.loads(soup.prettify())
coinlist = pd.DataFrame(dic['Data']).transpose()
coinlist.head()
```

```python
coinlist_id = coinlist['Id'].unique().tolist()
coinlist_id
```

```python
description = []
twitter = []
exc = {}

for i in coinlist_id:
    link = 'https://www.cryptocompare.com/api/data/coinsnapshotfullbyid/?id=%s' % i
    response = requests.get(link)
    data = response.json()
    
    try:
        description.append(data['Data']['General']['Description'])
        twitter.append(data['Data']['General']['Twitter'])

        exchanges = []
        markets = data['Data']['Subs']
        for item in markets:
            if item.split('~')[1] not in exchanges:
                exchanges.append(item.split('~')[1])
        exc[i] = exchanges
    except HTTPSConnectionPool:
        print i
coinlist['description'] = description
coinlist['twitter'] = twitter
coinlist['exchanges'] = exc
exc_df = pd.DataFrame(exc)
exc_df.head()
```
